```python
import powerlaw
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Path to degree distribution data
file_path = 'Data/Wastewater/Wastewater_cclasso_050_degree_distribution.csv'

# Load degree distribution data from csv
degree_distribution = pd.read_csv(file_path)
```

```python
fit = powerlaw.Fit(degree_distribution["Degree"], discrete=True)
```

```python
# Plot the data and the power law fit
fig = fit.plot_ccdf(linewidth=3, label='Empirical Data')
fit.power_law.plot_ccdf(ax=fig, color='r', linestyle='--', label='Power law fit')

# Add labels and title
fig.set_ylabel(u"p(X≥x)")
fig.set_xlabel("x")

# Rotate x-axis labels
ax = fig.axes
ax.tick_params(axis='x', rotation=90)

# Adjust the legend
handles, labels = fig.get_legend_handles_labels()
fig.legend(handles, labels, loc=3)

# Show the plot
plt.show()
```
